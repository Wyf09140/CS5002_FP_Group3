```python
# Libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas

# Modelling Libraries
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, minmax_scale, LabelEncoder
from sklearn.naive_bayes import CategoricalNB, GaussianNB, BernoulliNB
```

```python
# Import csv and convert to pandas
# Import csv and convert to pandas
filename = 'Ordered_Merged_Data_CDC_CALCAT.csv'
df = pandas.read_csv(filename)

print(df.columns)
```

```python
# Data Visualization - plot each parameter against the label
Y = df['hosp_yn']
X_cont = pandas.DataFrame(df['value'])
print(type(X_cont))
# Remove ICU - seems to be strongly correlated
X = df.drop(['hosp_yn', 'value', 'icu_yn', 'target_date'], axis = 1)

for param in X.columns:
    # Create heatmap for each of the parameters
    # Format for heatmap
    heatmapDat = pandas.crosstab(df[param], df['hosp_yn'])
    fig, ax = plt.subplots(figsize = (10,10))
    ax = sns.heatmap(heatmapDat, cmap = 'coolwarm')
    plt.savefig(f'{param}_plot.png', dpi = 300)
```

```python
# Data pre-processing - One-Hot Encoding - Input
X_enc = pandas.get_dummies(X, columns = X.columns)

# Normalize Reff Value
X_cont['Val_Norm'] = minmax_scale(X_cont, axis = 0)

# encode the Y label
le = LabelEncoder()
le.fit(Y)
Y_enc = le.transform(Y)
print(Y_enc)

# Combine X and X_cont - into dataframe for train-test split
X_processed = pandas.DataFrame(X_enc, columns = X_enc.columns)
X_processed['Value'] = X_cont["Val_Norm"]

print(np.unique(Y_enc, return_counts = True))
```

```python
# Split Data into train test
x_train, x_test, y_train, y_test = train_test_split(X_processed, Y_enc, stratify = Y_enc, random_state = 42)
```

```python
# SkLearn Naive Bayes Classifier Library
from sklearn.metrics import f1_score, confusion_matrix
clf = BernoulliNB()
model = clf.fit(x_train, y_train)
pred_y = model.predict(x_test)

# Compute F1 Score
print(f1_score(y_test, pred_y))
```
